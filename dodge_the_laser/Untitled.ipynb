```python
def approx(a,n):
    tri = a*n*(n+1)/2
    cor = -n/2
    return floor(tri+cor+frac(a))
```

```python
def big_tri(n):
    a = sqrt(2)
    N = floor(a*n)
    return N*(N+1)//2
```

```python
def specless(a,n):
    return ceil((n+1)/a)-1
```

```python
def triple(a,n):
    b = a/(a-1)
    N = floor(a*n)
    m = specless(b,N)
    a_approx = approx(a,n)
    b_approx = approx(b,m)
    bt = N*(N+1)//2
    return a_approx,b_approx,bt
```

```python
def root2_digit(n):
    if n==0:
        return 1
    if n==1:
        return 4
    W = 10**(n-1)
    X = 1
    Y = 0
    while min(X,Y)<W:
        X_temp = X
        X = X + 2*Y
        Y = X_temp + Y
    return floor(((((W-Y)*(X-Y))%Y)/Y)*10)
```

```python
def root2_int(n):
    lst = [str(root2_digit(n)) for n in range(n)]
    return int(''.join(lst))
```

```python
def big_n_times_root2(n):
    digits = [int(d) for d in str(n)]
    l = len(digits)
    root2 = root2_int(2*l)
    ans = 0
    for i,d in enumerate(digits):
        ans += d*root2
        root2 //= 10
    return ans // 10**l
```
